```python
# Threshold red color on the image
col_ref = [255,0,0]
thres_img = np.zeros(img.shape[0:2])
for r in range(img.shape[0]):
    for c in range(img.shape[1]):
        euc_dist = np.linalg.norm(img[r][c]-col_ref)
        thres_img[r][c] = euc_dist
```

```python
# Visualize red thresholded image
plt.imshow(thres_img<180)
```
